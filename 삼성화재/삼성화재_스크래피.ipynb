# 스크래피 데이터 불러오기

```python
import pandas as pd
```

```python
result=[]

for i in range(1,7):
    text = pd.read_csv('../뉴스크롤링_셀레니움/뉴스데이터csv/삼성화재_{}.csv'.format(i).format(i), dtype=str, sep=',', encoding='utf-8')
    result.append(text)
```

```python
total_raw_data = result[0].append(result[1]).append(result[2]).append(result[3]).append(result[4]).append(result[5]).append(result[5])
```

```python
total_raw_data
```

```python
total_raw_data=total_raw_data.reset_index(drop=True)
# reindex
```

```python
total_raw_data
```

```python
total_raw_data_2 = total_raw_data.drop(['author', 'category' ,'photo'], axis=1)
#필요없는 columns 제거
```

```python
total_raw_data_2
```

```python
media = total_raw_data_2['media']
content = total_raw_data_2['content']
```

```python
total_raw_data_3 = total_raw_data_2.drop(['media', 'content'], axis=1)
```

```python
total_raw_data_3
```

```python
total_raw_data_4 = pd.concat([media, total_raw_data_3], axis=1)
```

```python
total_raw_data_4
```

```python
total_raw_data_5 = pd.concat([total_raw_data_4, content], axis=1)
```

```python
total_raw_data_5
```

# 중복값, 결측값 제거

```python
total_data_dropna = total_raw_data_5.dropna()
```

```python
total_data_duplicate = total_data_dropna.drop_duplicates(['content'])
```

```python
total_data_duplicate.dropna()
```

```python
total_data_duplicate=total_data_duplicate.reset_index(drop=True)
```

```python
total_data_duplicate
```

```python
total_data_duplicate.columns = ['Media', 'Date', 'Title','Url', 'Content']
```

```python
total_data_duplicate
```

```python
total_data_duplicate.to_csv('../Data/삼성화재_스크래피.csv', encoding='utf-8', index= False)
```
