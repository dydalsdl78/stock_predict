# 셀레니움 데이터 불러오기

```python
import pandas as pd
```

```python
result=[]

for i in range(7,12):
    text = pd.read_csv('../뉴스크롤링_셀레니움/뉴스데이터csv/삼성화재_{}.csv'.format(i), dtype=str, sep=',', encoding='utf-8')
    result.append(text)
```

```python
total_raw_data = result[0].append(result[1]).append(result[2]).append(result[3]).append(result[4])
```

```python
total_raw_data
```

```python
total_raw_data=total_raw_data.reset_index(drop=True)
```

# 중복값, 결측값 제거

```python
total_data_dropna = total_raw_data.dropna()
```

```python
total_data_dropna
```

```python
total_data_dupliacate = total_data_dropna.drop_duplicates(['Content'])
```

```python
total_data_dupliacate
```

```python
total_data=total_data_dupliacate.reset_index(drop=True)
```

```python
total_data.to_csv('../Data/삼성화재_셀레니움.csv', encoding='utf-8', index=False)
```
