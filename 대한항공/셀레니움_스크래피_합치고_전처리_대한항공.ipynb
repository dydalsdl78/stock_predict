```python
for i in range(len(total_reindex['Content'])):
    try:
        

        total_reindex['Content'][i]= re.sub('(\w+\.)*\w+@(\w+\.)+[A-Za-z]+', '',total_reindex['Content'][i])
        
        if '(끝)' in total_reindex['Content'][i]:
            a = re.search('\(끝\)',total_reindex['Content'][i])
            total_reindex['Content'][i] = total_reindex['Content'][i][:a.start()]
          
            
        if  '*그림' in total_reindex['Content'][i]:     
            a = re.search('\*그림', total_reindex['Content'][i])
            total_reindex['Content'][i] =total_reindex['Content'][i][:a.start()] 
           
            
        if  '(서울=연합인포맥스)' in total_reindex['Content'][i]:
            a = re.search('\(서울\=연합인포맥스\)', total_reindex['Content'][i])
            total_reindex['Content'][i] =total_reindex['Content'][i][a.end():]
          
            
        if  '기자]' in total_reindex['Content'][i]:
            a = re.search('기자\]', total_reindex['Content'][i])
            total_reindex['Content'][i] =total_reindex['Content'][i][:a.start()]
           
            
        if  '기자 =' in total_reindex['Content'][i]:
            a = re.search('기자 =', total_reindex['Content'][i])
            total_reindex['Content'][i] =total_reindex['Content'][i][a.end():]
          
            
        if '사입니다]' in total_reindex['Content'][i]:
            a = re.search('사입니다\]', total_reindex['Content'][i])
            total_reindex['Content'][i] =total_reindex['Content'][i][a.end():]
           
            
        if '고충처리인 :' in total_reindex['Content'][i]:
            a = re.search('고충처리인 :', total_reindex['Content'][i])
            total_reindex['Content'][i] =total_reindex['Content'][i][:a.start()]    
            
            
        if '한경닷' in total_reindex['Content'][i]:
            a = re.search('한경닷', total_reindex['Content'][i])
            total_reindex['Content'][i] =total_reindex['Content'][i][:a.start()]  
           
            
        if '[정지성 ' in total_reindex['Content'][i]:
            a = re.search('\[정지성 :', total_reindex['Content'][i])
            total_reindex['Content'][i] =total_reindex['Content'][i][:a.start()]   
           
                
        if '[석민수 기자]' in total_reindex['Content'][i]:
            a = re.search('\[석민수 기자\]', total_reindex['Content'][i])
            total_reindex['Content'][i] =total_reindex['Content'][i][:a.start()]   
            
            
        if '[ⓒ 매일경제' in total_reindex['Content'][i]:
            a = re.search('\[\ⓒ 매일경제', total_reindex['Content'][i])
            total_reindex['Content'][i] =total_reindex['Content'][i][:a.start()]  
           
            
        if '인천/' in total_reindex['Content'][i]:
            a = re.search('인천\/', total_reindex['Content'][i])
            total_reindex['Content'][i] =total_reindex['Content'][i][:a.start()]  
            
            
        if '면 확대됩니다' in total_reindex['Content'][i]:
            a = re.search('면 확대됩니다', total_reindex['Content'][i])
            total_reindex['Content'][i] =total_reindex['Content'][i][a.end():]  
            
            
        if 'MTN 온라인 뉴스팀 =' in total_reindex['Content'][i]:
            a = re.search('MTN 온라인 뉴스팀', total_reindex['Content'][i])
            total_reindex['Content'][i] =total_reindex['Content'][i][:a.start()]  
        
        if '[매경' in total_reindex['Content'][i]:
            a = re.search('\[매경', total_reindex['Content'][i])
            total_reindex['Content'][i] =total_reindex['Content'][i][:a.start()] 
            
        if '기자 /' in total_reindex['Content'][i]:
            a = re.search('기자 //', total_reindex['Content'][i])
            total_reindex['Content'][i] =total_reindex['Content'][i][:a.start()] 
            
        if '세종=' in total_reindex['Content'][i]:
            a = re.search('세종\=', total_reindex['Content'][i])
            total_reindex['Content'][i] =total_reindex['Content'][i][:a.start()] 
        
        if '저작권자 ⓒ' in total_reindex['Content'][i]:
            a = re.search('저작권자 \ⓒ', total_reindex['Content'][i])
            total_reindex['Content'][i] =total_reindex['Content'][i][:a.start()] 
            
        if '\\저작권자 ©' in total_reindex['Content'][i]:
            a = re.search('\\저작권자 \©', total_reindex['Content'][i])
            total_reindex['Content'][i] =total_reindex['Content'][i][:a.start()] 
        
        if '당사의 허락' in total_reindex['Content'][i]:
            a = re.search('당사의 허락', total_reindex['Content'][i])
            total_reindex['Content'][i] =total_reindex['Content'][i][:a.start()] 
        
    except Exception as e:
        print(e)
    
```
