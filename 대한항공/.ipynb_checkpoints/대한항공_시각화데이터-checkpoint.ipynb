```python
import pandas as pd
from tqdm import tqdm_notebook
```

```python
text =  pd.read_csv('../Data/대한항공_Tone까지(최종분석데이터_빈도수제거x).csv', dtype=str, sep=',', encoding='utf-8')
```

```python
text
```

```python
T=text
```

```python
T['Tokens']
```

```python
for i in tqdm_notebook(range(len(T['Tokens']))):
    T['Tokens'][i] = T['Tokens'][i].replace(',', ' ')
```

```python
T_tokens = T['Tokens']
```

```python
T_tokens
```

```python
from gensim.summarization import keywords

keywords_text=[]

for i in tqdm_notebook(range(len(T_tokens))):
    keywords_text.append(keywords(T_tokens[i]).split('\n'))
```

```python
keywords_text
```

```python
keywords_first=[]
```

```python
for i in range(len(keywords_text)):
    keywords_first.append(keywords_text[i][0])
```
