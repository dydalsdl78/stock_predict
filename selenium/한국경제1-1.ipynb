```python
#2014/5/1 ~ 2016/5/1

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import WebDriverException,NoSuchElementException,StaleElementReferenceException,TimeoutException,ElementNotSelectableException,InvalidArgumentException,InvalidSelectorException,InvalidSwitchToTargetException,NoSuchAttributeException,NoSuchWindowException,UnexpectedTagNameException
import time
import random
import re


result_1 = []

options = Options()
options.add_argument("--no-sandbox")
options.add_argument("--disable-setuid-sandbox")
driver = webdriver.Chrome('C:/Users/user/kym/Project/chromedriver.exe', options=options)

url = "https://search.naver.com/search.naver?&where=news&query=%EB%8C%80%ED%95%9C%ED%95%AD%EA%B3%B5&sm=tab_pge&sort=2&photo=0&field=0&reporter_article=&pd=3&ds=2014.05.01&de=2016.05.01&docid=&nso=so:da,p:from20140501to20160501,a:all&mynews=1&start=0&refresh_start=0"
driver.get(url)

driver.find_element_by_xpath("""//*[@id="news_popup"]/a""").click()
driver.find_element_by_xpath("""//*[@id="ca_1015"]""").click()
driver.find_element_by_xpath("""//*[@id="_nx_option_media"]/div[2]/div[3]/button[1]/span""").click()
driver.find_element_by_xpath("""//*[@id="nx_option_mson"]""").click()

for j in range(0, 235):
    url = "https://search.naver.com/search.naver?&where=news&query=%EB%8C%80%ED%95%9C%ED%95%AD%EA%B3%B5&sm=tab_pge&sort=2&photo=0&field=0&reporter_article=&pd=3&ds=2014.05.01&de=2016.05.01&docid=&nso=so:da,p:from20140501to20160501,a:all&mynews=1&start={}1&refresh_start=0"
    driver.get(url.format(j))

    atag_list = driver.find_elements_by_xpath("""//*[@class="type01"]//dt/a""")
    date_list = driver.find_elements_by_xpath("""//*[@class="type01"]//dd[1]""")
    media_list = driver.find_elements_by_xpath("""//*[@class="type01"]//dd[1]/span[1]""")

    a_ls = []
    d_ls = []
    t_ls = []
    m_ls = []
    for i in range(0, len(atag_list)):
        a_ls.append(atag_list[i].get_attribute("href"))
        t_ls.append(atag_list[i].get_attribute("title"))
        d_ls.append(date_list[i].text)
        m_ls.append(media_list[i].text)

    for i in range(0, len(a_ls)):
        item_url = a_ls[i]
        item_date = d_ls[i]
        item_title = t_ls[i]
        item_media = m_ls[i]

        driver.get(item_url)
        r = random.randint(2, 3)
        time.sleep(r)

        p = re.compile('[0-9]+.[0-9]+.[0-9]+')
        date_ls = p.findall(item_date)
        date = ''.join(date_ls)

        try:
            Content_path = driver.find_element_by_xpath("""//*[@id="articletxt"]""")
            hangul = re.compile('[^ --a-zA-Z0-9.]+')
            content = hangul.sub('', Content_path.text)

            result_1.append([item_media, date, item_title, item_url, content])

            driver.back()
            r = random.randint(2, 3)
            time.sleep(r)

        except (WebDriverException, NoSuchElementException, StaleElementReferenceException, TimeoutException, ElementNotSelectableException, InvalidArgumentException, InvalidSelectorException, InvalidSwitchToTargetException, NoSuchAttributeException, NoSuchWindowException, UnexpectedTagNameException) as e:
            driver.back()
            print(e)
            pass
```
